# Аналіз виживання на Титаніку: Дослідницький проєкт

Команда проєкту:
* Королишин Марія — попередня обробка даних та побудова моделі "Дерево рішень".
* Арбайтер Віта — побудова моделі "Логістична регресія".

## 1. Вступ та мета дослідження
У цьому проєкті проводиться порівняльний аналіз двох алгоритмів машинного навчання для прогнозування ймовірності виживання пасажирів лайнера "Титанік". 

На етапі підготовки даних було виконано такі кроки:
* Заповнення пропущених значень базовими статистичними метриками (вік та вартість квитка).
* Конвертація категоріальних змінних у числовий формат.
* Конструювання нових ознак на основі існуючих.

## 2. Результати моделі "Дерево рішень"
Алгоритм Decision Tree (максимальна глибина = 4) дозволив візуалізувати логіку класифікації у вигляді ієрархічної структури. Нижче наведено згенерований граф моделі, який відображає критерії розбиття на кожному вузлі.

![Граф Дерева рішень](decision_tree_final.png)

## 3. Результати моделі "Логістична регресія"
Алгоритм Logistic Regression дозволив оцінити лінійний вплив кожної ознаки на ймовірність виживання. На графіку нижче наведено значення коефіцієнтів моделі для кожного фактора.

![Важливість факторів](logistic_regression_importance.png)

## 4. Висновки
Обидві моделі були успішно натреновані на спільному наборі даних. Результати збережено у файли submission_decision_tree.csv та submission_logistic_regression.csv. Структура проєкту відповідає вимогам модульної розробки.
